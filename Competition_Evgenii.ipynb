```python
import numpy as np
import pandas as pd
from dateutil import parser
from unidecode import unidecode
import seaborn as sns
import datetime

pd.set_option('display.max_columns', 116)
```

```python
train = pd.read_csv('train.csv',sep=';')
```

```python
test = pd.read_csv('test.csv',sep=',')
```

```python
train['date'] = train['date'].apply(lambda x: pd.to_datetime(x,dayfirst=True,format='%d.%m.%Y'))
```

```python
regions_list = train['region'].sort_values().unique()
```

```python
for num_reg in regions_list:
    min_date = train[train['region'] == num_reg]['date'].min().date()
    max_date = train[train['region'] == num_reg]['date'].max().date()
    print('Номер региона: {}. История продаж с {} по {}'.format(num_reg, min_date, max_date))
```

```python
# Признаки с разными типами данных
columns_list = ['pasta',
 'legumes',
 'bread',
 'flour',
 'rice',
 'groats',
 'potatoes',
 'сucumbers_tomatoes',
 'vegetables',
 'roots',
 'сabbage',
 'fruit',
 'sugar',
 'candy',
 'biscuits',
 'mutton',
 'beef',
 'chicken',
 'pork',
 'fish',
 'herring',
 'curd',
 'sour_creame',
 'butter',
 'milk_kefir',
 'cheese',
 'egg',
 'margarine',
 'oil',
 'salt',
 'spice',
 'tea',
 'cpi_1',
 'cpi_2',
 'cpi_3',
 'pasta_value',
 'legumes_value',
 'bread_value',
 'flour_value',
 'rice_value',
 'groats_value',
 'potatoes_value',
 'сucumbers_tomatoes_value',
 'vegetables_value',
 'roots_value',
 'сabbage_value',
 'fruit_value',
 'sugar_value',
 'candy_value',
 'biscuits_value',
 'mutton_value',
 'beef_value',
 'chicken_value',
 'pork_value',
 'fish_value',
 'herring_value',
 'curd_value',
 'sour_creame_value',
 'butter_value',
 'milk_kefir_value',
 'cheese_value',
 'egg_value',
 'margarine_value',
 'oil_value',
 'salt_value',
 'spice_value',
 'tea_value',
 'ai92',
 'ai95',
 'ai98',
 'dt',
 'ai92_value',
 'ai95_value',
 'ai98_value',
 'dt_value']
```

```python
for column in columns_list:
    print(column,'-',train[column].map(type).unique())
```

```python
# декодируем странные string значения меняем ',' на '.' и убираем пробелы для возможности привести к float
for column in columns_list:
    train[column] = train[column].apply(lambda x: unidecode(x).replace(",",'.').replace(' ','') if type(x) == str else x)
```

```python
for column in columns_list:
    try:
        train[column] = train[column].astype(np.float32)
    except:
        print(column)
```

```python
# Округляем значения
for column in columns_list:
    try:
        train[column] = train[column].apply(lambda x: round(x,2))
    except:
        print(column)
```

```python
for num_reg in regions_list:
    days = train[train['region'] == num_reg].shape[0]
    print('Номер региона: {}. Количество дней в истории продаж {}'.format(num_reg, days ))
```

```python
# В регионе 21 дублируются продажи с неправильным OKATO. оставляем данные с правильным OKATO
train = train[~((train['region'] == 21)&(train['okato'] == 1000000000))]
```

```python
skiped_cols_in54 = []
for column in train.columns:
    skiped = train[column].isna().sum()
    if skiped > 0:
        skiped_cols_in54.append(column)
        
    print('Столбец: {}. Количество пропусков {}'.format(column, skiped ))
```

```python
for column in skiped_cols_in54:
    train[column].fillna(train[(train['region'] == 54)&
                           (train['date']>pd.to_datetime('2019-04-23'))&
                           (train['date']<pd.to_datetime('2019-04-28'))][column].mean(),inplace=True)
```
